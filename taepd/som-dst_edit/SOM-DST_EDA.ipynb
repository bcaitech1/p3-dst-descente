```python
import json
```

```python
pred_dict27 = json.load(open('preds_27.json'))
pred_dict28 = json.load(open('preds_28.json'))
```

```python
def create_miss_pred_dict(pred_dict):
    miss_pred_dict = {}
    for k, v in pred_dict.items():
        if v[0] != v[1]:
            miss_pred_dict[k] = v
    return miss_pred_dict
```

```python
miss_pred_dict27 = create_miss_pred_dict(pred_dict27)
miss_pred_dict28 = create_miss_pred_dict(pred_dict28)
```

```python
print(len(pred_dict27))

print(len(miss_pred_dict27))
print(len(miss_pred_dict28))
```
